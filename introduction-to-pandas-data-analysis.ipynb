### Subsetting Data

```python
df.columns
```

```python
df = df[['title', 'description', 'publishTime', 
    'duration_seconds',
    'viewCount', 'likeCount', 'commentCount']]
df
```

### Subsetting rows

```python
# Subsetting using loc
df.loc[df['viewCount'] > 1_000_000]
```

```python
# Subsetting using query
df_subset2 = df.query('viewCount > 1000000')
```

```python
df_subset2
```

```python
df.loc[df['viewCount'].isna()]
```

```python
df = df.loc[~df['viewCount'].isna()]
df
```

### Casting dtypes

```python
df.dtypes
```

```python
df.loc[df['likeCount'].isna()]
```

```python
df['likeCount'] = df['likeCount'].fillna(0)
```

```python
df.loc[df['likeCount'].isna()]
```

```python
df['viewCount'] = df['viewCount'].astype('int')
df['likeCount'] = df['likeCount'].astype('int')
```

```python
df.dtypes
```

```python
df['publishTime'] = pd.to_datetime(df['publishTime'])
```

```python
df['publishTime']
```

```python
pd.to_numeric(df['likeCount'].astype('str'))
```
